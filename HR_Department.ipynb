```python
x_cat = df[['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus']]
x_cat.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
x_cat = ohe.fit_transform(x_cat).toarray()
display(x_cat.shape)
```

```python
display(x_cat)
```

```python
x_cat = pd.DataFrame(x_cat)
x_cat.head()
```

```python
x_num = df[['Age', 'DailyRate', 'DistanceFromHome',	'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement',	'JobLevel',	
            'JobSatisfaction',	'MonthlyIncome',	'MonthlyRate',	'NumCompaniesWorked',	'OverTime',	'PercentSalaryHike', 'PerformanceRating',	
            'RelationshipSatisfaction',	'StockOptionLevel',	'TotalWorkingYears'	,'TrainingTimesLastYear'	, 'WorkLifeBalance',	'YearsAtCompany',
            'YearsInCurrentRole', 'YearsSinceLastPromotion',	'YearsWithCurrManager']]
x_num.head()
```

```python
x_all = pd.concat([x_cat,x_num], axis=1)
x_all.head()
```

```python
x_all.columns = x_all.columns.astype(str)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
x = mms.fit_transform(x_all)
x
```

```python
x.shape
```

```python
y = df['Attrition'].values
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state=1)
display(x_train.shape)
display(x_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lgr = LogisticRegression()
lgr.fit(x_train, y_train)
```

```python
y_pred = lgr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
y_rfc = rfc.predict(x_test)
print(f"Accuracy {100*accuracy_score(y_test,y_rfc)}%")
print(confusion_matrix(y_test,y_rfc))
print(classification_report(y_test,y_rfc))
```

```python
y_rfc_train = rfc.predict(x_train)
print(f"Accuracy {100*accuracy_score(y_train,y_rfc_train)}%")
print(confusion_matrix(y_train,y_rfc_train))
print(classification_report(y_train,y_rfc_train))
```

```python
from xgboost import XGBClassifier
xgbc = XGBClassifier()
xgbc.fit(x_train, y_train)
y_tsxgb = xgbc.predict(x_test)
print(f"Accuracy {100*accuracy_score(y_test,y_tsxgb)}%")
print(confusion_matrix(y_test,y_tsxgb))
print(classification_report(y_test,y_tsxgb))
```
